```python
import pandas as pd
import numpy as np

df = pd.read_csv("UCI_Credit_Card.csv")
df.head()
```

```python
df = df.rename(columns={'default.payment.next.month': 'Target'})
```

```python
df['default.payment.next.month'].value_counts()
```

```python
def drop_correlated_features(df, target, threshold=0.9):
  
    corr = df.corr().abs()
    target_corr = corr[target].drop(target)

    upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
    
    to_drop = set()

    for col in upper.columns:
        for row in upper.index:
            value = upper.loc[row, col]
            if not np.isnan(value) and value > threshold:
                
                if target_corr[row] < target_corr[col]:
                    to_drop.add(row)
                else:
                    to_drop.add(col)

    return list(to_drop)

cols_to_drop = drop_correlated_features(df, target="default.payment.next.month", threshold=0.85)

print("will drop columns", cols_to_drop)

df = df.drop(columns=cols_to_drop)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier

models = {
    'LogisticRegression': LogisticRegression(max_iter=500, random_state=42),
    'GaussianNB': GaussianNB(),
    'KNN': KNeighborsClassifier(),
    'SVM': SVC(random_state=42),   # no probability=True -> much faster for CV
    'MLP': MLPClassifier(max_iter=500, random_state=42)
}
```
